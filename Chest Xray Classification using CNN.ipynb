```python
import numpy as np
```

```python
#preprocessing

def loadImages(path, urls, target): #This line defines a function named loadImages that takes three parameters: path, urls, and target.
  # Here, two empty lists images and labels are initialized. These lists will store the loaded images and their corresponding labels.
  images = []
  labels = []
  for i in range(len(urls)): # The code iterates over each element in the urls list. len(urls) gives the total number of URLs, and range(len(urls)) generates the indices from 0 to len(urls) - 1.

    img_path = path + "/" + urls[i] # For each URL, it constructs the full path to the image by concatenating the path argument with the current URL.

    img = cv2.imread(img_path) # It reads the image from the specified file path using OpenCV's imread function and assigns it to the variable img. This function returns a numpy array representing the image.

    img = img / 255.0   # This line normalizes the pixel values of the image by dividing each pixel value by 255.0. This step ensures that pixel values are in the range [0, 1], which is commonly done to aid model training.

    # It resizes the image to a fixed size of 100x100 pixels using OpenCV's resize function:
    img = cv2.resize(img, (100, 100))

    # The normalized and resized image (img) is appended to the images list, and the corresponding label (target) is appended to the labels list.
    images.append(img)
    labels.append(target)

    # After processing all images, the list of images (images) is converted to a numpy array using np.asarray(). This conversion is done to facilitate further processing since many machine learning libraries work efficiently with numpy arrays.
  images = np.asarray(images)

  #Finally, the function returns a tuple containing the numpy array of images (images) and the list of labels (labels). This allows the caller to access both the images and their corresponding labels after loading.
  return images, labels
```

```python
covid_path = "COVID-19_Radiography_Dataset/COVID/images"
covidUrls = os.listdir("COVID-19_Radiography_Dataset/COVID/images")
covidImages, covidTarget = loadImages(covid_path, covidUrls, 1)
```

```python
normal_path = "COVID-19_Radiography_Dataset/Normal/images"
normalUrls = os.listdir("COVID-19_Radiography_Dataset/Normal/images")
normalImages, normalTarget = loadImages(normal_path, normalUrls, 0)
```

1 = covid, 0 = normal

```python
# converting into numpy array because a list cannot be used to find the shape of data
covidImages = np.asarray(covidImages)
normalImages = np.asarray(normalImages)
```

```python
covidImages.shape
```

```python
normalImages.shape
```

When you use np.r_, you're essentially concatenating the arrays covidImages and normalImages along the first axis, resulting in a single NumPy array. This operation effectively combines the data from both datasets into one array.

For example, if covidImages has a shape of (N1, H, W, C) and normalImages has a shape of (N2, H, W, C), where N1 and N2 are the number of images in the COVID and normal datasets, respectively, and H, W, and C are the height, width, and number of channels (e.g., RGB), then the resulting data array will have a shape of (N1 + N2, H, W, C).

```python
data = np.r_[covidImages, normalImages]
```

```python
data.shape
```

```python
targets = np.r_[covidTarget, normalTarget]
```

```python
targets.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, targets, test_size = 0.25)
```

```python
#!pip install tensorflow
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization # 6 layers
```

```python
model = Sequential([
  Conv2D(32, (3, 3), input_shape=(100, 100, 3), activation="relu", padding="same"),
  MaxPooling2D(pool_size=(2, 2)),
  BatchNormalization(),
  
  Conv2D(64, (3, 3), activation="relu", padding="same"),
  MaxPooling2D(pool_size=(2, 2)),
  BatchNormalization(),
  
  Conv2D(128, (3, 3), activation="relu", padding="same"),
  MaxPooling2D(pool_size=(2, 2)),
  BatchNormalization(),
  
  Flatten(),
  Dense(128, activation="relu"),
  Dropout(0.5),
  Dense(1, activation="sigmoid")
])
```

filters parameter specifies the number of filters/kernels to be applied in the convolution operation. Each filter is responsible for detecting different features in the input data.

kernel_size parameter defines the size of the convolutional window. It specifies the height and width of the 2D convolutional window. Typically, the kernel size is set as a tuple (height, width). Common choices include (3, 3) and (5, 5). These sizes are small enough to capture local patterns while being computationally efficient.

In a CNN, when a filter/kernel is applied to the input image, it starts from the top-left corner and slides over the image, computing dot products between the filter weights and the corresponding input pixels. Padding refers to adding extra pixels (usually zeros) around the input image to control the size of the output feature maps.

There are two common types of padding:

Valid Padding:

In valid padding, no padding is applied to the input image. As the filter/kernel slides over the input image, it only computes the dot product where the filter and input fully overlap. With valid padding, the output feature map will be smaller than the input image, especially when using large kernel sizes. Same Padding:

In same padding, enough padding is added to the input image so that the output feature map has the same spatial dimensions as the input. Specifically, if the filter size is F, the padding added to each side of the input image is (F - 1) / 2. Same padding helps in preserving spatial information and can be particularly useful in deeper layers of the network.

```python
model.compile(optimizer='adam', loss=tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,batch_size=32,epochs=10,validation_data=(x_test, y_test))
```

```python
plt.plot(model.history.history["accuracy"], label = "train accuracy")
plt.plot(model.history.history["val_accuracy"], label = "test accuracy")
plt.legend()
plt.show()
```
